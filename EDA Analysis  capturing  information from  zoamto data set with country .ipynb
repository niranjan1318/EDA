```python
import plotly.express as px

fig = px.histogram(
    df,
    x="Aggregate rating",
    nbins=20,
    title="Interactive Distribution of Restaurant Ratings",
    hover_data=df.columns
)

fig.show()
```

```python
fig = px.box(
    df,
    x="Price range",
    y="Aggregate rating",
    title="Price Range vs Aggregate Rating"
)

fig.show()
```

```python
fig = px.violin(
    df,
    x="Price range",
    y="Aggregate rating",
    box=True,
    points="all",
    title="Violin Plot: Price Range vs Rating"
)

fig.show()
```

```python
fig = px.box(
    df,
    x="Price range",
    y="Votes",
    title="Price Range vs Votes"
)

fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numerical_df = df[["Aggregate rating", "Votes", "Price range"]]

plt.figure(figsize=(8,6))
sns.heatmap(
    numerical_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Heatmap")
plt.show()
```
